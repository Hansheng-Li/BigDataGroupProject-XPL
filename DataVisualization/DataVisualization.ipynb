#### Each State - Employment

```python
## Data cleaning and reformat
def clean_num(row):
    row = row.replace(",", "")
    return row

df_unemploy["Initial Claims"] = df_unemploy["Initial Claims"].map(clean_num)
df_unemploy["Continued Claims"] = df_unemploy["Continued Claims"].map(clean_num)
df_unemploy["Covered Employment"] = df_unemploy["Covered Employment"].map(clean_num)
```

```python
State_List = df_unemploy.State.unique()

## Date
df_unemploy["Filed week ended"] = pd.to_datetime(df_unemploy["Filed week ended"])
df_unemploy["Reflecting Week Ended"] = pd.to_datetime(df_unemploy["Reflecting Week Ended"])

## Numeric
df_unemploy["Covered Employment"] = pd.to_numeric(df_unemploy["Covered Employment"])
df_unemploy["Insured Unemployment Rate"] = pd.to_numeric(df_unemploy["Insured Unemployment Rate"])

df_unemploy["Initial Claims"] = pd.to_numeric(df_unemploy["Initial Claims"])
df_unemploy["Continued Claims"] = pd.to_numeric(df_unemploy["Continued Claims"])
```

```python
df_unemploy_2020 = df_unemploy[df_unemploy["Filed week ended"].dt.year==2020]
df_unemploy_2021 = df_unemploy[df_unemploy["Filed week ended"].dt.year==2021]
```

```python
df_unemploy.State.unique()
```

```python
## choose based on 2020 GDP rank for each state
Rich_State_List = ["California","Texas","New York"]
Poor_State_list = ["Vermont", "Wyoming", "Alaska"]

State_list = Rich_State_List + Poor_State_list
State_list
```

```python
df_unemploy_2020_chozenState = df_unemploy_2020[df_unemploy_2020.State.isin(State_list)]
df_unemploy_2021_chozenState = df_unemploy_2021[df_unemploy_2021.State.isin(State_list)]
```

```python
df_unemploy_2020_chozenState['Month'] = df_unemploy_2020_chozenState["Filed week ended"].map(lambda x: x.month)
```

```python
df_unemploy_2020_chozenState.head()
```

```python
df_unemploy_2020_chozenState_simple = pd.DataFrame(columns = ["State","Month","meanUnemplymentRatePerMonth"])

for _state in df_unemploy_2020_chozenState.State.unique():
    for _month in df_unemploy_2020_chozenState[df_unemploy_2020_chozenState.State==_state].Month.unique():
        State = _state
        Month = _month
        meanUnemplymentRatePerMonth = df_unemploy_2020_chozenState[(df_unemploy_2020_chozenState.Month==_month) & (df_unemploy_2020_chozenState.State==_state)]["Insured Unemployment Rate"].mean()
        df_unemploy_2020_chozenState_simple.loc[len(df_unemploy_2020_chozenState_simple.index)] = [State, Month, meanUnemplymentRatePerMonth] 
```

```python
df_unemploy_2020_chozenState_pivot = df_unemploy_2020_chozenState_simple.pivot(
                                                        index = 'Month',
                                                        columns = 'State',
                                                        values = 'meanUnemplymentRatePerMonth')

fig, ax7 = plt.subplots(figsize=(10,7))
ax7 = df_unemploy_2020_chozenState_pivot.plot(xticks=df_unemploy_2020_chozenState_pivot.index, 
                                              ax = ax7,
                                              title = "UnemployRate/Month in each State")
ylab = ax7.set_ylabel('average unemploy rate')
```
